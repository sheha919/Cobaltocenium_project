```python
df_mod = df.sort_values('Std Proton Afinity', ascending= True)
df_mod
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(data=df_mod, x='Substituent', y='Std Proton Afinity',marker='v', 
              markersize=10, linewidth = 1, linestyle = '--', color = 'r')
plt.xticks(rotation=90)
plt.ylabel('Standardized Proton Affinity')
plt.xlabel('')
plt.savefig('std_proton_afinity.png');
```
